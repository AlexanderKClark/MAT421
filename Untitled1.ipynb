## !7.4 Lagrange Polynomial Interpolation
Lagrange polynomial interpolation finds a single polynomial that goes through all data points, unlike the piecewise cubic functions that cubic spline interpolation uses. This polynomial is called a Lagrange polynomial, $L(x)$, and it is a linear combination of Lagrange basis polynomuals, $P_i(x)$, where

\begin{equation*}
P_i(x)=\prod\limits_{j=1,j\neq i}^n\frac{x-x_i}{x_i-x_j}
\end{equation*}

and

\begin{equation*}
L(x)=\sum\limits_{i=1}^ny_iP_i(x).
\end{equation*}

By construction, $P_i(x_j)=1$ when $i=j$ and $P_i(x_j)=0$ when $i \neq j$. Therefore, $L(x_i)=y_i$ for every point.

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')

x = [0, 1, 2]
y = [1, 3, 2]
P1_coeff = [1,-1.5,.5]
P2_coeff = [0, 2,-1]
P3_coeff = [0,-.5,.5]
P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)

x_new = np.arange(-1.0, 3.1, 0.1)

plt.figure
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Lagrange Basis Polynomials')
plt.grid
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
L = P1 + 3*P2 + 2*P3
fig = plt.figure()
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Lagrange interpolation is defined outside of the interval $[x_1,x_n]$, and it will grow quickly and unbounded when outside of this interval. As such, it is inadvisable to use this to interpolate outside of the interval.

```python
plt.style.use('seaborn-v0_8-whitegrid')
from scipy.interpolate import lagrange

f = lagrange(x, y)

fig = plt.figure()
plt.plot(x_new, f(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 17.5 Newton Polynomial Interpolation
The general form of a Newton polynomial of $n-1$ order going through $n$ points is:
\begin{equation*}
f(x)=a_0+a_1(x-x_0)+a_2(x-x_0)(x-x_1)+\cdots+a_n(x-x_0)(x-x_1)\cdots(x-x_n)=\sum\limits_{i-0}^na_in_i(x)
\end{equation*}
where $n_i(x)=\prod_{j=0}^{i-1}(x-x_j)$. The coefficients $a_i$ can be determined algorithmically. Consider the following:

$\begin{aligned}
f(x_0)& =a_0 \\
f(x_1)& = a_0+a_1(x_1-x_0)
\end{aligned}\implies a_1=\frac{y_1-y_0}{x_1-x_0}$

$\hphantom{\begin{aligned}
f(x_0)& =a_0 \\
f(x_1)& = a_0+a_1(x_1-x_0)
\end{aligned}}\implies$ $a_2=$$\displaystyle\frac{\frac{y_2-y_1}{x_2-x_1}-\frac{y_1-y_0}{x_1-x_0}}{x_2-x_1}$

$\hphantom{\begin{aligned}
f(x_0)& =a_0 \\
f(x_1)& = a_0+a_1(x_1-x_0)
\end{aligned}\implies=}\vdots$


So, if $f[x_1,x_0]$ is defined as $\frac{y_1-y_0}{x_1-x_0}$, then $f[x_2,x_1,x_0]=$$\displaystyle\frac{\frac{y_2-y_1}{x_2-x_1}-\frac{y_1-y_0}{x_1-x_0}}{x_2-x_0}=\frac{f[x_2,x_1]-f[x_1-x_0]}{x_2-x_0}$

and so,
\begin{equation*}
f[x_k,x_{k-1},\dots,x_1,x_0]=\frac{f[x_k,x_{k-1},\dots,x_2,x_1]-f[x_{k-1},x_{k-2},\dots,x_1,x_0]}{x_k-x_0}.
\end{equation*}


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline

def divided_diff(x, y):
  #function to calculate the divided differences table
  n = len(y)
  coef = np.zeros([n, n])
  # the first column is y
  coef[:,0] = y
  for j in range(1,n):
    for i in range(n-j):
      coef[i][j] = \
      (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
  return coef
def newton_poly(coef, x_data, x):
  #evaluate the newton polynomial at x
  n = len(x_data) - 1
  p = coef[n]
  for k in range(1,n+1):
    p = coef[n-k] + (x -x_data[n-k])*p
  return p

x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])
# get the divided difference coef
a_s = divided_diff(x, y)[0, :]
# evaluate on new data points
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)
plt.figure()
plt.plot(x, y, 'bo')
plt.plot(x_new, y_new)
```
