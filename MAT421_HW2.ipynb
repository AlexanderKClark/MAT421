## 17.1 Interpolation Problem Statement
Interpolation is the process of determining new data points froms a discrete set of original points. Formally, this means we want an estimation function $\hat{y}(x)$ such that $\hat{y}(x_i)=y_i$ for every point in the data set. Furthermore, given a new $x^*$, the output of the interpolation function $\hat{y}(x^*)$ is an estimate for the function value at $x^*$.

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')

x = [1, 2, 3, 4, 5]
y = [1, 8, 6, 3, 2]

plt.plot(x, y, 'o')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Five Arbitrary Points')
plt.xticks(x)
plt.yticks(y)

# Display the plot
plt.show()
```

##17.2 Linear Interpolation
With linear interpolation, the estimated point is assumed to be incident with the line connecting the points on the left and on the right. Under the assumption WLOG that $x_i<x_{i+1}$ and $x_i<x<x_{i+1}$, then the linear interpolation of $x$ is

 $\hat{y}=y_i+\frac{\left(y_{i+1}-y_i\right)\left(x-x_i\right)}{x_{i+1}-x_i}$.



```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 8, 6, 3, 2]
point = 3.5
f = interp1d(x, y)
y_hat = f(point)
print(y_hat)
```

```python
plt.plot(x, y, '-ob')
plt.plot(point, y_hat, 'ro')
plt.title('Linear Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(x + [point])
plt.yticks(y + [y_hat])
plt.show()
```

## 17.3 Cubic Spline Interpolation
With cubic spline interpolation, the estimated point is assumed to be incident with the curve of a cubic function connecting the points on the left and the right. These cubic functions are naturally of the form

$S_i(x)=a_ix^3 + b_ix^2 + c_ix + d$,

where $x$ is not a point from the discrete set and $i=1,\dots,n-1$. For $n$ points, there are $n-1$ functions and $4(n-1)$ unknowns. This is solved by setting up the appropriate system of equations.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 8, 6, 3, 2]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(1, 5, 100)
y_new = f(x_new)
point = 3.5
interpolation = f(point)
print(interpolation)
```

```python
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'bo')
plt.plot(point, interpolation, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(x + [point])
plt.yticks(y+[interpolation])
plt.show()
```
